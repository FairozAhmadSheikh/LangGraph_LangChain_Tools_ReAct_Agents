# **First Graph**

# Step 1 : Define Schema

```python
# Define Schema 
"""
Schema are of Two types : 
LIGHT WEIGHT :  Means this does not have strict rules
HEAVY WEIGHT :  This has strict rules to follow 
"""

from typing import TypedDict,List

class graph_schema(TypedDict):
    name:str
    message:str
```

# Step 2 : Build Node or Function of a node 


```python
def welcome(schema:graph_schema) -> graph_schema:
    curr_name=schema['name']
    curr_message=schema['message']
    response=llm.invoke(f'My name is {curr_name} and i want to say {curr_message}').content
    schema['message']=f'Your Message was {curr_message} and LLM Response is {response}'
    return schema
```

# Step 3 : Define State Graph 

```python

from langgraph.graph import StateGraph,START,END 
graph=StateGraph(state_schema=graph_schema)

# add nodes to the graph
graph.add_node('welcome',welcome)

# add edges to the graph
graph.add_edge(START,'welcome')
graph.add_edge('welcome',END)
```

# Step 4 : Compile the model  

```python

from IPython.display import Image

first_graph=graph.compile()

Image(first_graph.get_graph().draw_mermaid_png())
```

# Step5 : Run the Graph 


```python
first_graph.invoke({
    'name':'Fairoz',
    'message':'How Are You dooing ?'})
```
