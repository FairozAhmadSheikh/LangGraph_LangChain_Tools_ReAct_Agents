```python
from langchain_google_genai import ChatGoogleGenerativeAI

from dotenv import find_dotenv,load_dotenv

import os 

if os.environ["GOOGLE_API_KEY"]:
    print("key found sucess")
else:
    print("Key not found ")
```

```python
llm=ChatGoogleGenerativeAI(model='gemini-2.5-flash-lite')
```

```python
from langchain_core.prompts  import ChatPromptTemplate

my_prompt=ChatPromptTemplate.from_messages(
    [
        ('system',"You are helpfull social media content expert helping content creators to create engaging posts "),
        ("human","Create a Social Media Post regarding {topic}")
        
    ])
filled_prompt=my_prompt.invoke({'topic':"The importance of apple in health "})

print(filled_prompt)
```

```python
#  Create a chain Instead 
chain= my_prompt | llm
response=chain.invoke({"topic":"why is batch size 32 in ML ?"})
```

```python
print(response.content)
```
