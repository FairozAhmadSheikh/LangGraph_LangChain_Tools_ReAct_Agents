# Example to create a Tool

```python
from langchain_community.tools import DuckDuckGoSearchRun

duck_search=DuckDuckGoSearchRun()

response=duck_search.invoke("What is an Large Language Model ?")
```

```python
print(response)
```

```python
# Tool For Research Paper

from langchain_community.tools import ArxivQueryRun
from langchain_community.utilities import ArxivAPIWrapper

arxiv_querry=ArxivQueryRun(api_wrapper=ArxivAPIWrapper())

results=arxiv_querry.invoke("Explainable AI ")
```

```python
print(results)
```

```python
# Wikipedia Tool
from langchain_community.tools import WikipediaQueryRun
from langchain_community.utilities import WikipediaAPIWrapper

wiki_search=WikipediaQueryRun(api_wrapper=WikipediaAPIWrapper())

res=wiki_search.invoke("Who is Omar Abdullah?")
```

```python
print(res)
```
