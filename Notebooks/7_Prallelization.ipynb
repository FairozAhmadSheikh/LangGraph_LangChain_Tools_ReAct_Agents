# Workflow : **Parellization**

>We have a person that has to create a post parallely for three social media 
* linkedIn , 
* Insta 
* Twitter 

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from dotenv import load_dotenv,find_dotenv
import os 
if os.environ['GOOGLE_API_KEY']:
    print(f'API Key Success')
else:
    raise ValueError("Please Provide an API key First")

load_dotenv(find_dotenv(),override=True)
```

```python
llm=ChatGoogleGenerativeAI(model='gemini-2.5-flash-lite')
```

```python
# Lets Create a Schema 
from typing import TypedDict,List

class graph_schema(TypedDict):
      topic:str
      linkedin:str
      insta:str
      twitter:str
```

```python
# Create Node Functions

# Create post for Instagram
def create_post_insta(state:graph_schema)->graph_schema:
    topic=state['topic']
    post=llm.invoke(f'Create a Instagram Post for {topic}. Keep the tone casual and engaging ').content
    state['insta']=post

    return {'insta':post}

# Create Post for LinkedIn
def create_post_linkedin(state:graph_schema)->graph_schema:
    topic=state['topic']
    post=llm.invoke(f'Create a LinkedIn Post for {topic}. Keep the tone professional and informative').content
    state['linkedin']=post

    return {'linkedin':post}

# Create post for Twitter
def create_post_twitter(state:graph_schema)->graph_schema:
    topic=state['topic']
    post=llm.invoke(f'Create a Instagram Post for {topic}. Keep the tone quick ').content
    state['twitter']=post

    return {'twitter':post}
```
